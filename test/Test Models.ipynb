### Description

This notebook is for testing to see what models work best with our TinderBCI.

```python
%matplotlib inline
# imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# models
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

# PCA
from sklearn.decomposition import KernelPCA
from sklearn.decomposition import PCA

# preprocessing
from sklearn.preprocessing import LabelEncoder

# metrics
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score

# model selection
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import train_test_split
```

```python
# Read in the data
random_data_df = pd.read_csv("./data/sample_data.csv", sep=",")
random_data_df.head(10)
```

```python
# Grab the labels
y_labels = random_data_df.pop('left/right').values
y_labels_plot = ['r' if c == 'N' else 'b' for c in y_labels]

# Encode them
le = LabelEncoder()
encoded = le.fit_transform(y_labels)
```

```python
# check on the dataframe. we should have two columns and a shape of (20000, 2)
print(random_data_df.columns, random_data_df.shape)
```

```python
# make sure the label encoder worked
print(encoded, le.inverse_transform(encoded))
```

```python
# create training and testing data
X_train, X_test, y_train, y_test = train_test_split(random_data_df.values, encoded)
```

```python
# Plot the data thus far. Pretty useless but we'll give it a go anyway
plt.scatter(random_data_df['attention'], random_data_df['meditation'], c=y_labels_plot)
```

```python
param_grid = [{'solver': ['lbfgs', 'sag']}] # TODO: Add more params to this grid
```

```python
log_reg = LogisticRegression()
grid_search = GridSearchCV(log_reg, param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X_train, y_train)
```

```python
print(grid_search.best_params_, "\n", grid_search.best_estimator_)
```

```python
results = grid_search.best_estimator_.predict(X_test)
f1_score(y_test, results)
```
