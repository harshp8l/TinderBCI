# Testing Notebook

### Description

This notebook is for testing to see what models work best with our TinderBCI.

```python
%matplotlib inline

# imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# models
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import LinearSVC

# PCA
from sklearn.decomposition import PCA

# preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler

# metrics
from sklearn.metrics import accuracy_score

#not sure
from sklearn.manifold import TSNE

# model selection
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split

#exporting
from sklearn.externals import joblib

from collections import Counter
from mpl_toolkits import mplot3d
```

```python
# Read in the data
EEG_data = pd.read_csv("./data/output-1.csv", sep=",", header=None)
EEG_data.columns = ['rawValue', 'attention', 'meditation', 'delta', 'theta', 'lowAlpha', 'highAlpha', 'lowBeta', 'highBeta', 'lowGamma', 'midGamma', 'std_rawValue', 'std_attention', 'std_meditation', 'std_delta', 'std_theta', 'std_lowAlpha', 'std_highAlpha', 'std_lowBeta', 'std_highBeta', 'std_lowGamma', 'std_midGamma', 'choice']
EEG_data = EEG_data[EEG_data['meditation'] != 0]

real_data = EEG_data.filter(['attention','meditation','choice'],axis=1)
real_data.head(10)
```

```python
# grab labels
y_labels = real_data.pop('choice').values
y_labels_plot = ['r' if c == 'right' else 'b' for c in y_labels]
fig = pd.plotting.scatter_matrix(real_data, alpha=0.2)
```

```python
plt.scatter(real_data['attention'][:45], real_data['meditation'][:45], c=y_labels_plot)
plt.show()
```

```python
X = StandardScaler().fit_transform(real_data)
```

```python
Counter(y_labels).keys() # equals to list(set(words))
Counter(y_labels).values() 
```

```python
# Encode the labels
le = LabelEncoder()
encoded = le.fit_transform(y_labels)

X_train, X_test, y_train, y_test = train_test_split(real_data.values, encoded)
```

```python
# make sure the label encoder worked
print(encoded, le.inverse_transform(encoded))
```

```python
X_tsne = TSNE(n_components=3).fit_transform(X)
plt.scatter(X_tsne[:,0],X_tsne[:,1], c=y_labels_plot)
plt.show()
```

```python
X_tsne = TSNE(n_components=3).fit_transform(X)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d') 
ax.scatter3D(X_tsne[:,0],X_tsne[:,1], X_tsne[:,2],color=y_labels_plot)
```

```python
scaler = MinMaxScaler()

X_train, X_test, y_train, y_test = train_test_split(real_data.values, encoded)
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

## RandomForest

```python
clf = RandomForestClassifier(n_estimators=100, max_depth=2, random_state=0)
clf.fit(X_train_scaled, y_train)
```

```python
y_train_pred = clf.predict(X_train_scaled)
```

```python
accuracy_score(y_train_pred, y_train)
```

```python
y_test_pred = clf.predict(X_test_scaled)
#y_test_pred = [1 if num == 0 else 0 for num in y_test_pred]
```

```python
accuracy_score(y_test_pred, y_test)
```

```python
X_train_scaled
```

## SVM

```python
param_grid = [{'loss': ['hinge', 'squared_hinge']}]
svc_clf = LinearSVC()
grid_search = GridSearchCV(svc_clf,param_grid,cv=5,n_jobs=-1,scoring='accuracy')
grid_search.fit(X_train_scaled, y_train)
```

```python
print(grid_search.best_params_, "\n", grid_search.best_estimator_)
```

```python
results = grid_search.best_estimator_.predict(X_test_scaled)
#results = [1 if num == 0 else 0 for num in results]
```

```python
accuracy_score(y_test,results)
```
